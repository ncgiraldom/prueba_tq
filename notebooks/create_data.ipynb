```python
df = pd.DataFrame(datos)
```

```python
# Rango de fechas
fechas = pd.date_range(start='2018-01-01', end='2024-12-01', freq='MS')
```

```python
# Crear un nuevo DataFrame para los resultados
df_resultados = pd.DataFrame({'Fecha': fechas})
```

```python
# Simulación de series de tiempo
def generar_serie(promedio):
    # Tendencia (puedes ajustarla)
    tendencia = np.linspace(0, promedio * 0.2, len(fechas))
    # Estacionalidad (ciclos anuales)
    estacionalidad = promedio * 0.1 * np.sin(2 * np.pi * fechas.dayofyear / 365.25)
  
    # Ruido (aleatorio)
    ruido = np.random.normal(0, promedio * 0.1, len(fechas))
  
    # Suma de componentes
    serie = promedio + tendencia + estacionalidad + ruido

    # Redondear a enteros
    serie = np.round(serie).astype(int)

    # Asegurar valores no negativos
    serie = np.maximum(0, serie)

    return serie
```

```python
# Generar datos para cada SKU y guardarlos en el nuevo DataFrame
for sku, promedio in zip(df['SKU'], df['Promedio']):
    serie = generar_serie(promedio)
    df_resultados[sku] = serie
```

```python
# Guardar datos en formato CSV
df_resultados.to_csv('datos_series_tiempo.csv', index=False)
print("Datos generados y guardados en datos_series_tiempo.csv")
```
